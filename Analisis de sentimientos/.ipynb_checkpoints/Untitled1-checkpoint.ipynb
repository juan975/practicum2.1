```python
import pandas as pd
```

```python
#Importacion de datos de etrenamiento
df = pd.read_excel('data_entrenada_propia_GP.ods', engine='odf')
df
```

```python
print(df.isnull().sum())
```

```python
# Eliminar filas con valores nulos en 'feedback' o 'polaridad'
df = df.dropna(subset=['feedback', 'polaridad'])
```

```python
print(df.isnull().sum())
```

```python
# Tomamos los datos solo de texto y sus etiquetas
X = df['feedback']
y = df['polaridad']
```

```python
# Revision de la columna 'polaridad'
print(df['polaridad'].value_counts())
```

```python
df.loc[(df['polaridad'] == 'NEUTRO') | (df['polaridad'] == 'NUE') | (df['polaridad'] == 'NOE')]
```

```python
df['polaridad'] = df['polaridad'].replace({
    'NEUTRO': 'P',
    'NOE': 'NONE',
    'NUE': 'NEU'
})
```

```python
# Revision de los cambios en la columna 'polaridad'
print(df['polaridad'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Configuracion para el entrenamiento del modelo
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)#Test_size hace que se entrene el modelo con el 80% y se pruebe con el 20%
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
#Limpieza de datos con lematizacion
def cleaner_text (text):
    if not isinstance(text, str):  # Si no es cadena, devolver vacío
        return ""
    
    # Tokenizar el texto
    tokens = word_tokenize(text.lower())

    # Quitar stop words de las frases
    filtered_tokens = [token for token in tokens if token not in stopwords.words('spanish')]

    # Lematizar las palabras
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

    # Join the tokens back into a string
    cleaned_text = ' '.join(lemmatized_tokens)
    return cleaned_text
```

```python
# Aplicar la limpieza de texto a los conjuntos
X_train = X_train.apply(cleaner_text)
X_test = X_test.apply(cleaner_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)
```

```python
# Realizar predicciones
y_pred = model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred))
```
